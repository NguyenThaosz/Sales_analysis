TASK 1: SET UP ENVIRONMENT AND LOAD DATA

First, import all library we need:

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

Import data from CSV files and Merger 12-month data

```python
path = 'E:\PROJECT\Sale analysist/'
frames = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path  + file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)
df = result
```

```python
df
```

ADD 'Month' column

```python
df['Month'] = df['Order Date'].str[0:2]
df
```

```python
set(df['Month'])
```

Remove 'nan' and 'or' value

```python
# remove 'nan' value
df = df.dropna(how = 'all')
df
```

```python
# Find out why the value 'Or' appears in the merger data table
df[df['Month']== 'Or']
```

```python
# get rid of 'Or' value
df = df[df['Month'] != 'Or']
df
```

```python
# Recheck unique value in 'Month' column
set(df['Month'])
```

TASK 3: REPORTING

3.1: What was the best month for sales ?

```python
# Check schema of the dataset.
df.info()
```

```python
# covert to numeric
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast = 'integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast = 'float')
```

```python
# Add 'Sales' column to calculate the value of each order ID
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df
```

```python
#change 'Sales' column position
moving_col = df.pop('Sales')
df.insert(4,'Sales',moving_col)
df
```

```python
# calculate 'Sales' column value by month
sales_value = df.groupby('Month').sum()['Sales']
sales_value
```

```python
# The month with the highest sales is 12
sales_value.max()
```

```python
# Showing by bar chart
month = range(1,13)
plt.bar(x= month, height = sales_value )
plt.xticks(month)
plt.xlabel('Month')
plt.ylabel('Sale in USD')
plt.show()
```

3.2: What city has the best Sales ?

```python
# Add 'City' column
address_to_city = lambda address:address.split(',')[1]
df['City'] = df['Purchase Address'].apply(address_to_city)
df
```

```python
# calculate 'Sales' column value by city
sales_value_city = df.groupby('City').sum()['Sales']
sales_value_city
```

```python
# The city with the highest sales is San Francisco
sales_value_city.max()
```

```python
cities = []

for city, sales in sales_value_city.items():
    cities.append(city)
print(cities)
```

```python
# Showing bar chart
plt.bar(x = cities, height = sales_value_city)
plt.xticks(cities, rotation = 45, size = 9)
plt.xlabel('Cities')
plt.ylabel('Sales in USD')
```

3.3: What time should we display ads to maximize the likelihood of customer's buying product ?

```python
# covert 'Order date' column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
print(df['Order Date'].dtypes)
```

```python
df['Hours'] =  df['Order Date'].dt.hour
df
```

```python
# calculate order ID by hour
order_value_hour = df.groupby('Hours').count()['Sales']
order_value_hour
```

```python
# Showing plot chart
times = []
for time, sale in order_value_hour.items():
    times.append(time)
plt.plot(times, order_value_hour)
plt.grid()
plt.xticks(times)
plt.xlabel('Hours')
plt.ylabel('Sales in USD')
plt.show()
```

```python
# The time frames with the high number of orders is from 11 to 13 and 18 to 20.
```
